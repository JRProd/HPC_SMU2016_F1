```python
import matplotlib.pyplot as plt
```

```python
x = open("x.txt",'r').read()
x.replace(" ","")
x = x.split("\n")
x.remove("")

f = open("f.txt",'r').read()
f.replace(" ","")
f = f.split("\n")
f.remove("")

p = open("p.txt",'r').read()
p.replace(" ","")
p = p.split("\n")
p.remove("")

for i in range(0,len(x)-1):
    x[i] = float(x[i].strip())
    f[i] = float(f[i].strip())
    p[i] = float(p[i].strip())
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

ax.plot(x,f,"k*", label="$f(x)$")
ax.plot(x,p,"r-", label="Polynomial degree 4")
ax.set_title("$f(x)=e^x$ vs $T_4(x)$, $T_8(x)$, & $T_{12}(x)$")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

legendPlot1 = ax.legend(loc="upper left", shadow=True)

plt.show()
```
