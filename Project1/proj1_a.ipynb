 # Taylor Series and Floating-Point Error

Project 1 consist of two parts. Part 1 uses Taylor Polynomials to approximate a function and show the error in approximation based on the degree of the Taylor Polynomial used. Part 2 I calculate the relative error and the upper bounds of that error in approximating a suficiently smooth fuction's derivative.

## Approximation of a Function by Taylor Polynomials

To begin part 1, I implemnted Horner's nested multiplication using C++. This reduced the number of arithmatic opperations performed for each of out Taylor Polynomial approximations.
I reduced $$P = a_0 + a_1 x + a_2 x^2 + ... + a_{n-1} x^{n-1} + a_n x^n$$ to $$P = a_0 + (a_1 + x(a_2 + x^2(...+ x^{n-1}(a_{n-1} + a_n x^n)...)))$$ This reduced the number of opperations performed to n additions and n multiplications for every time an approximation is computed.
Next, I created three varying degrees of a taylor series representation of $f(x) = e^x$. The first taylor polynomial was$$T_4(x) = 1 + 1x + \frac{1}{2}x^2 + \frac{1}{6}x^3$$
The second taylor polynomial was$$T_8(x) = 1 + 1x + \frac{1}{2}x^2 + \frac{1}{6}x^3 + \frac{1}{24}x^4 + \frac{1}{120}x^5 + \frac{1}{720}x^6 + \frac{1}{5040}x^7$$
The third and final taylor polynomial wass$$T_{12}(x) = 1 + 1x + \frac{1}{2}x^2 + \frac{1}{6}x^3 + \frac{1}{24}x^4 + \frac{1}{120}x^5 + \frac{1}{720}x^6 + \frac{1}{5040}x^7 + \frac{1}{40320}x^8 + \frac{1}{367880}x^9 + \frac{1}{3678800}x^{10} + \frac{1}{39916800}x^{11}$$
For each of these taylor polynomials I approximated every x value in the vector z = -1, 10.99, -0.98, ..., -0.01, 0, 0.01, ..., 0.98, 0.99, 1.0 and created the vectors **p4, p8, p12** respectfully. They contained the approximation of $f(x) = e^x$ of each taylor seires for each x value in z.
Next, I create **err4, err8, err12** that contain$$|err_n - p_n|$$ or the absolute error of the approximation from the real value. The final portion of Part A required I write **p4, p8, p12, err4, err8, err12, & f** (f being the $f(x) = e^x$ for every x in z) to .txt files.

```python
z = open("z.txt",'r').read()
z.replace(" ","")
z = z.split("\n")
z.remove("")

p4 = open("p4.txt",'r').read()
p4.replace(" ","")
p4 = p4.split("\n")
p4.remove("")

p8 = open("p8.txt",'r').read()
p8.replace(" ","")
p8 = p8.split("\n")
p8.remove("")

p12 = open("p12.txt",'r').read()
p12.replace("  ","")
p12 = p12.split("\n")
p12.remove("")

fun = open("f.txt",'r').read()
fun.replace("  ","")
fun = fun.split("\n")
fun.remove("")

err4 = open("err4.txt",'r').read()
err4.replace("  ","")
err4 = err4.split("\n")
err4.remove("")

err8 = open("err8.txt",'r').read()
err8.replace("  ","")
err8 = err8.split("\n")
err8.remove("")

err12= open("err12.txt",'r').read()
err12.replace("  ","")
err12 = err12.split("\n")
err12.remove("")

for i in range(0,len(fun)-1):
    z[i] = float(z[i].strip())
    p4[i] = float(p4[i].strip())
    p8[i] = float(p8[i].strip())
    p12[i] = float(p12[i].strip())
    fun[i] = float(fun[i].strip())
    err4[i] = float(err4[i].strip())
    err8[i] = float(err8[i].strip())
    err12[i] = float(err12[i].strip())
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

ax.plot(z,fun,"k-", label="$f(x)=e^x$")
ax.plot(z,p4,"r-", label="Taylor Polynomial degree 4")
ax.plot(z,p8,"b-", label="Taylor Polynomial degree 8")
ax.plot(z,p12,"g-", label="Taylor Polynomial degree 12")
ax.set_title("$f(x)=e^x$ vs $T_4(x)$, $T_8(x)$, & $T_{12}(x)$")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

legendPlot1 = ax.legend(loc="upper left", shadow=True)

plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

ax.plot(z,err4,"g-", label="Error of 4th degree Taylor Polynomial")
ax.plot(z,err8,"r-", label="Error of 8th degree Taylor Polynomial")
ax.plot(z,err12,"b--", label="Error of 12th degree Taylor Polynomial")
ax.set_title("Absolute Error between $f(x)=e^x$ and $T_4(x)$, $T_8(x)$, & $T_{12}(x)$")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

legendPlot2 = ax.legend(loc="upper left", shadow=True)

plt.show()
```

## Errors in a Forward Finite Diffrence Approximation

```python
n = open("n.txt",'r').read()
n = n.split("\n")
n.remove("")

h = open("h.txt",'r').read()
h = h.split("\n")
h.remove("")

r = open("r.txt",'r').read()
r = r.split("\n")
r.remove("")


R = open("R.txt",'r').read()
R = R.split("\n")
R.remove("")
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

ax.semilogy(n, r, "b-", label="Relative Error")
ax.semilogy(n, R, "r--", label="Upper Bound of Error")
ax.set_title("N vs Relative Error and Upper Bound of Error")
ax.set_xlabel("N")
ax.set_ylabel("Relative Error of the Derivative Approximation")

legendPlot3 = ax.legend(loc="lower right", shadow=True)
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

ax.loglog(h, r, "b-", label="Relative Error")
ax.loglog(h, R, "r--", label="Upper Bound of Error")
ax.set_title("H vs Relative Error and Upper Bound of Error")
ax.set_xlabel("H")
ax.set_ylabel("Relative Error of the Derivative Approximation")

legendPlot4 = ax.legend(loc="lower right", shadow=True)
```
